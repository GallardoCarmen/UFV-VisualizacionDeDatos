```python
df = pd.read_csv('datos_ejercicio_ventas.csv', delimiter = ',')
df.head()
```

Cosas a tener en cuenta relevantes:
- ventas acutales y predichas
- horizonte de previsión
- número de paises productos...
- historico de datos actuals/forecast
- Forecast distintos
- rango de cantidades

```python
ventas_reales = df[df['SCENARIO'] == 'actual'].shape[0]
predicciones = df[df['SCENARIO'] == 'AI_forecast'].shape[0]

print(f'Ventas reales: {ventas_reales}')
print(f'Predicciones: {predicciones}')
```

```python
horizonte = df[ (df['COUNTRY'] == 'Portugal') & (df['SUBBRAND'] == 'Lipton (L3)') & (df['FORECAST'] == 'AI_P02F')].shape[0]

print(f'Horizonte de predicción: {horizonte}')
```

```python
paises = df['COUNTRY'].nunique()
productos = df['SUBBRAND'].nunique()

print(f'Número de países únicos: {paises}')
print(f'Número de productos únicos: {productos}')
```

```python
fig = px.histogram(df, x="COUNTRY", color = "COUNTRY")
fig.show()
```

```python
fig = px.histogram(df, x="COUNTRY", color="COUNTRY", title='Cantidad de registros por país', labels={'COUNTRY':'País', 'count':'Cantidad de registros'},
                  category_orders={"COUNTRY": df['COUNTRY'].value_counts().index})
fig.show()
```

```python
fig = px.histogram(df, x="SUBBRAND", color="SUBBRAND", title='Cantidad de registros por sub-marca', labels={'SUBBRAND':'Sub-marca', 'count':'Cantidad de registros'},
                category_orders={"SUBBRAND": df['SUBBRAND'].value_counts().index})

fig.show()
```
